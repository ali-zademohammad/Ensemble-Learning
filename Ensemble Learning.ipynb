## Decision Tree

### Load  library and data

```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor # Import Decision Tree Classifier
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
housing = fetch_california_housing(as_frame=True)
df = housing.frame
# df.to_csv('california_housing_prices.csv', index=False)

X = df.drop('MedHouseVal',axis=1)
y = df['MedHouseVal']
# MedInc: میانگین درآمد منطقه.
# HouseAge: سن میانگین خانه‌ها در منطقه.
# AveRooms: میانگین تعداد اتاق‌ها در هر خانه.
# AveBedrms: میانگین تعداد اتاق‌خواب‌ها در هر خانه.
# Population: جمعیت منطقه.
# AveOccup: میانگین تعداد افراد ساکن در هر خانه.
# Latitude: عرض جغرافیایی.
# Longitude: طول جغرافیایی.
# MedHouseVal: مقدار میانه قیمت خانه‌ها در منطقه (هدف برای مدل‌ها)
```

```python
# Normal
# scaler = StandardScaler()
# X = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # 70% training and 30% test
# Create Decision Tree classifer object
clR = DecisionTreeRegressor(criterion="squared_error", max_depth=5)

# Train Decision Tree Classifer
clR = clR.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clR.predict(X_test)
print("Accuracy:",mean_squared_error(y_test, y_pred))
```

```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
# ترسیم درخت تصمیم
plt.figure(figsize=(20, 15))  # تنظیم اندازه شکل
plot_tree(clR, filled=True, feature_names=housing.feature_names, rounded=True)
plt.title("Decision Tree for California Housing Prices")
plt.savefig("decision_tree.png", dpi=300)  # dpi=300 برای کیفیت بالا

plt.show()  # نمایش درخت
```
